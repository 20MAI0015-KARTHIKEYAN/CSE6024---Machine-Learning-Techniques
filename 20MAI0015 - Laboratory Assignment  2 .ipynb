4. Display the Month with most of loans issued date 

```python
data['date_issued_month'] = pd.to_datetime(data['date_issued'])
data['date_issued_month'].dt.month.value_counts(0)
```

5. Display the count of ‘Teacher’ who are ‘Owners’ 

```python
count=0
for c,d in zip(data['employment'],data['householder']):
  if c=='Teacher' and d=='OWNER':
      count=count+1
print("count of ‘Teacher’ who are ‘Owners’ is",count)
```

6. Display the ‘Employment’ of customers who mostly ‘Rent'

```python
lst=[]
for a,b in zip(data['householder'],data['employment']):
  if a=='RENT':
     lst.append(b)
def countX(lst, x): 
    return lst.count(x)
elements = list(set(lst))

for i in range (0, len(elements)):
    print('{} Occurs {} times'.format(elements[i], countX(lst, elements[i])))
```
